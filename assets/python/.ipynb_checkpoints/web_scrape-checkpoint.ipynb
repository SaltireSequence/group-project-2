# Scraping Code 

## Scraping Simply Recipe Website Index Links

```python
import requests
from bs4 import BeautifulSoup
import pprint
from scraper_api import ScraperAPIClient
from splinter import Browser
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# array of dicts structure to hold returned keys and values.
recipe_links = {}

res = requests.get("https://www.simplyrecipes.com/")
soup = BeautifulSoup(res.text,"html.parser")
```

```python
for div in soup.find("div", class_="rnav-menus").find_all("div", class_="rnav-menu"):
    recipe_type = div.find("span").get_text(strip=True)
    recipe_links[recipe_type] = [i.find("a")["href"] for i in div.find_all("li")]
```

```python
pprint.pprint(recipe_links)
```

```python
# Creating a DataFrame of links
link_db = pd.DataFrame.from_dict(recipe_links, orient='index')
link_db
```

```python
# Transposing columns and rows appropriately.
df2_transposed = link_db.T
df2_transposed.head()
```

## Scraping Recipe Page for Ingredients List

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path)
```

```python
from scraper_api import ScraperAPIClient
client = ScraperAPIClient('41e7f420344eb4f9359fb9360d4d4708')
result = client.get(url = 'https://www.simplyrecipes.com/recipes/cooking_for_two_strawberry_almond_oat_smoothie/')
```

```python
# Initiating HTML parser, using BeautifulSoup
soup = BeautifulSoup(result.text, "html.parser")
```

```python
out, current = {}, ''
for tag in soup.select('.recipe-callout h2, .ingredient'):
    if tag.name == 'h2':
        current = tag.text + " " + "Ingredients"
        out[current] = []
    else:
        out[current].append(tag.text)

from pprint import pprint
pprint(out, width = 120)
```
